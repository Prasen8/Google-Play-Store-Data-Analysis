```python
# 1) Clean the 'Reviews' data and change the type from 'object' to 'float'
# Convert all entries in the 'Reviews' column to strings
reviews = [str(i) for i in apps_df['Reviews']]

def clean_reviews(reviews_list):
    """
    This function replaces 'M' with million in the reviews data
    and converts the values to float.
    """
    cleaned_data = []
    for review in reviews_list:
        if 'M' in review:
            review = review.replace('M', '')
            review = float(review) * 1000000  # 1M = 1,000,000
        else:
            # If it's already a number (without 'M'), convert it to float
            review = float(review)
        cleaned_data.append(review)
    return cleaned_data

# Apply the cleaning function
apps_df['Reviews'] = clean_reviews(reviews)

# Now you can convert the 'Reviews' column to float
apps_df['Reviews'] = apps_df['Reviews'].astype(float)
```

```python
# 1) Clean the 'Size' data and change the type from 'object' to 'float'

# Remove uncertain values
index = apps_df[apps_df['Size'] == '1,000+'].index
apps_df.drop(axis=0, inplace=True, index=index)

# Convert all entries in the 'Size' column to strings
sizes = [str(i) for i in apps_df['Size']]

def clean_sizes(sizes_list):
    """
    This function removes 'M' and 'k' from sizes,
    converting 'k' (kilobytes) into megabytes.
    """
    cleaned_data = []
    for size in sizes_list:
        if 'M' in size:
            size = size.replace('M', '')
            size = float(size)  # Convert to float
        elif 'k' in size:
            size = size.replace('k', '')
            size = float(size)
            size = size / 1024  # Convert kilobytes to megabytes
        elif 'Varies with device' in size:
            size = float(0)  # Represent 'Varies with device' with value 0
        else:
            # If it's already a float (or a string that can be converted to float), convert it
            size = float(size)
        cleaned_data.append(size)
    return cleaned_data

# Apply the cleaning function
apps_df['Size'] = clean_sizes(sizes)

# Now you can convert the 'Size' column to float
apps_df['Size'] = apps_df['Size'].astype(float)
```

```python
# 1) Clean the 'Installs' data and change the type from 'object' to 'float'

# Convert all entries in the 'Installs' column to strings
installs = [str(i) for i in apps_df['Installs']]

def clean_installs(installs_list):
    cleaned_data = []
    for install in installs_list:
        # Remove commas and plus signs
        if ',' in install:
            install = install.replace(',', '')
        if '+' in install:
            install = install.replace('+', '')
        
        # Convert to float
        try:
            install = float(install)  # Convert to float directly
        except ValueError:
            install = 0.0  # Handle any unexpected values by assigning 0.0
        
        cleaned_data.append(install)
    return cleaned_data

# Apply the cleaning function
apps_df['Installs'] = clean_installs(installs)

# Now you can convert the 'Installs' column to float (if not already done)
apps_df['Installs'] = apps_df['Installs'].astype(float)
```

```python
# 1) Clean the 'Price' data and change the type from 'object' to 'float'

# Convert all entries in the 'Price' column to strings
prices = [str(i) for i in apps_df['Price']]

def clean_prices(prices_list):
    cleaned_data = []
    for price in prices_list:
        # Remove the dollar sign
        if '$' in price:
            price = price.replace('$', '')
        
        # Convert to float, handling any unexpected values
        try:
            price = float(price)  # Convert to float directly
        except ValueError:
            price = 0.0  # Handle any unexpected values by assigning 0.0
        
        cleaned_data.append(price)
    return cleaned_data

# Apply the cleaning function
apps_df['Price'] = clean_prices(prices)

# Now you can convert the 'Price' column to float (if not already done)
apps_df['Price'] = apps_df['Price'].astype(float)
```

```python
def replace_with_median(series):
    """
    Given a series, replace the rows with null values 
    with median values
    """
    return series.fillna(series.median())

apps_df['Rating'] = apps_df['Rating'].transform(replace_with_median)
apps_df['Rating'] = apps_df['Rating'].astype(float)
```

```python
# remove the record where 'Type' is having null value
index = apps_df[apps_df['Type'].isna()].index
apps_df.drop(axis=0, inplace=True, index=index)
```

```python
# check on null values
apps_df.isna().sum()
```

```python
# grouping the data starting with App and Reviews
apps_df = apps_df.groupby(['App', 'Reviews', 'Category', 'Rating', 'Size', 'Installs', 'Type', 'Price', 'Content Rating', 'Genres', 
                           'Last Updated', 'Current Ver', 'Android Ver'], as_index=False)
# reassign Installs values with their mean
apps_df = apps_df['Installs'].mean()
# sort the dataframe by Reviews descendingly
apps_df.sort_values(by='Reviews', ascending=False, inplace=True)
# drop duplicate rows based on App 
apps_df.drop_duplicates(subset=['App'], inplace=True)
apps_df
```

```python
# get the number of apps for each category
sns.set_style('darkgrid')
plt.figure(figsize=(10, 5))
sns.countplot(x='Category', data=apps_df)
plt.title('Number of Apps Per Category')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.show()
```

```python
# get the number of installs for each category
categories = apps_df.groupby('Category')
category_installs_sum_df = categories[['Installs']].sum()
category_installs_sum_df = category_installs_sum_df.reset_index()  # to convert groupby object into dataframe

plt.figure(figsize=(10, 5))
sns.barplot(x='Category', y='Installs', data=category_installs_sum_df)
plt.xticks(rotation=90)
plt.ylabel('Installs (e+10)')
plt.title('Number of Installs For Each Category')
plt.show()
```

```python
# Distribution of App Ratings
plt.figure(figsize=(10, 5))
sns.histplot(apps_df['Rating'], bins=20, kde=True, color='blue')
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie chart for Free vs Paid apps
type_counts = apps_df['Type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=140)
plt.title('Proportion of Free vs Paid Apps')
plt.show()
```

```python
# Bar Graph: Top 10 Most Installed Apps
top_installed = apps_df.nlargest(10, 'Installs')
plt.figure(figsize=(10, 5))
sns.barplot(y=top_installed['App'], x=top_installed['Installs'], palette='coolwarm')
plt.xlabel('Installs')
plt.ylabel('App')
plt.title('Top 10 Most Installed Apps')
plt.show()
```

```python
# Pie Chart: Distribution of Content Ratings
content_rating_counts = apps_df['Content Rating'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(content_rating_counts, labels=content_rating_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=140)
plt.title('Distribution of Content Ratings')
plt.show()
```

```python
# Additional Visualization: Category-wise Average Rating
category_ratings = apps_df.groupby('Category')['Rating'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Rating', y='Category', data=category_ratings.sort_values('Rating', ascending=False), palette='viridis')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.title('Average Rating by Category')
plt.show()
```
